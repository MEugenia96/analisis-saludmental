### 5. Descripción de las escalas clínicas utilizadas

#### Estas puntuaciones derivan de dos test:

* _PARA LA ANSIEDAD_: Test denominado GAD-7 (Generalized Anxiety Disorder-7). El cual evalúa la gravedad de los síntomas de ansiedad si bien no es un diagnóstico definitivo ha demostrado ser un instrumento confiable y válido. Su escala va de 0 a 21, dependiendo de las respuesta a 7 preguntas.
  
<div style="margin-left:20px; text-align:left;">
  <table style="border-collapse: collapse; text-align:left;">
    <tr><th>PUNTAJE TOTAL</th><th>NIVEL DE ANSIEDAD</th></tr>
    <tr><td>0 - 4</td><td>Mínimo o ninguno</td></tr>
    <tr><td>5 - 9</td><td>Leve</td></tr>
    <tr><td>10 - 14</td><td>Moderado</td></tr>
    <tr><td>15 - 21</td><td>Severo</td></tr>
  </table>
</div>

  
  
* _PARA LA DEPRESIÓN_: Test PHQ-9 – Cuestionario de Salud del Paciente (Patient Health Questionnaire-9), es una herramienta de evaluación clínica diseñada para identificar la presencia y severidad de síntomas asociados a la depresión mayor. Este cuestionario consta de nueve ítems y su escala va de 0 a 27-30.

 <div style="margin-left:20px; text-align:left;">
  <table style="border-collapse: collapse; text-align:left;">
    <tr><th>PUNTAJE TOTAL</th><th>NIVEL DE DEPRESIÓN</th></tr>
    <tr><td>0 - 4</td><td>Mínimo o ninguno</td></tr>
    <tr><td>5 - 9</td><td>Leve</td></tr>
    <tr><td>10 - 14</td><td>Moderado</td></tr>
    <tr><td>15 - 19</td><td>Moderadamente severo</td></tr>
    <tr><td>20 - 27</td><td>Severo</td></tr>
  </table>
</div>



### 6. Clasificación por niveles

#### De acuerdo a los rangos declarados en la tabla anterior realicé _"Creación de variables categóricas"_.

#### *ANSIEDAD OBSERVADA - Según test GAD-7 – Trastorno de Ansiedad Generalizada.*

```python
# Creación de rangos categóricos.

rangos = [0, 4, 9, 14, 21]  
categorias = ["Mínimo o ninguno", "Leve", "Moderado", "Severo"]

df['anxiety_level'] = pd.cut(df["anxiety_score"], bins=rangos, labels=categorias, include_lowest=True)

```

```python
## Gráfico

plt.figure(figsize=(5,5))
sns.countplot(x='anxiety_level', data=df, hue='anxiety_level', palette='Set2', legend=False)
```

```python
## Porcentaje por cada categoría.

df['anxiety_level'].value_counts(normalize=True) * 100
```

#### *DEPRESIÓN OBSERVADA - Según test PHQ-9 – Cuestionario de Salud del Paciente.*

```python
# Creación de rangos categóricos.

rangosD = [0,4,9,14,19,30]
categoriasD = ['Mínimo o ninguno','Leve','Moderado','Moderadamente severo','Severo']

df['depression_level'] = pd.cut(df["depression_score"], bins=rangosD, labels=categoriasD, include_lowest=True)
```

```python
## Gráfico

plt.figure(figsize=(8,5))
sns.countplot(x='depression_level', data=df, hue='depression_level', palette='Set2', legend=False)
```

```python
## Porcentaje por cada categoría.

df['depression_level'].value_counts(normalize=True) * 100
```

#### 7. Relación entre ansiedad y depresión

```python
tabla = pd.crosstab(df['anxiety_level'], df['depression_level'])
```

```python
sns.heatmap(tabla, annot=True, cmap='coolwarm', fmt='d')
plt.title('Relación entre niveles de ansiedad y depresión')
plt.xlabel('anxiety_level')
plt.ylabel('depression_level')
plt.show()
```

#### 9. Factores asociados a altos niveles

#### Edad_________________________________________

##### *Agrupación por rangos*

```python
rangosAge = [0, 25, 35, 45, 55, 65]  
categoriasAge = ["18-24 años", "25-34 años", "35-44 años", "45-54 años", "55-65 años"]

df['age_range'] = pd.cut(df['age'], bins=rangosAge, labels=categoriasAge, include_lowest=True)
```

##### Los datos muestran que la ansiedad promedio se mantiene en niveles **moderados** (entre 10 y 11 puntos) en todos los rangos etarios. 
##### Sin embargo:

* Hay un aumento progresivo desde los 18 hasta los 44 años.
* El pico máximo se alcanza entre los 45 y 54 años.
* Luego, en el grupo de 55 a 65 años, se observa una leve disminución, aunque los niveles siguen siendo moderados.

#### RELACION DEPRESIÓN - EDAD

```python
sns.lineplot(x='age_range', y='depression_score', data=df, color='forestgreen')
```

##### Los datos muestran que la depresión promedio se mantiene en niveles **Moderadamente severo** (entre 14,8 y 15,6 puntos) en todos los rangos etarios. 
##### Sin embargo:

* El pico máximo se alcanza entre los 18 y 24 años.
* El mínimo se obtiene en el rango de los 25-34 años.
* Luego, sube, manteniendose en lo moderadamente severo desde los 35 a 65 años.


#### Género_________________________

```python
totales = df['gender'].value_counts()
```

#### RELACIÓN ANSIEDAD - GENERO

##### Para este análisis, se consideran únicamente los casos con puntuaciones superiores a 9, correspondientes a niveles moderados o severos de ansiedad.

```python
filtro1 = df[df['anxiety_score'] > 9]
filtro1

ansiedad = filtro1['gender'].value_counts()
```

```python
##Cuánto del total por género sufre ansiedad severa o moderada

porcentaje = (ansiedad / totales * 100).round(2)
porcentaje
```

```python
sns.violinplot(x='gender', y='anxiety_score', hue = 'gender', data=filtro1, palette = 'Set2', legend=False)
```


##### A partir del análisis se observa que más del 50% de las personas en cada género presentan niveles de **ansiedad severa o moderada**.
#### Los porcentajes son muy similares entre los distintos grupos (oscilan entre ~54% y ~58%).

#### RELACIÓN DEPRESION - GENERO

##### Para este análisis, se consideran únicamente los casos con puntuaciones superiores a 10, correspondientes a niveles moderados o severos de depresión.

```python
filtro2 = df[df['depression_score'] > 10]
filtro2

depresion = filtro2['gender'].value_counts()
```

```python
##Cuánto del total por género sufre depresion severa o moderada

porcentaje2 = (depresion / totales *100).round(2)
porcentaje2
```

```python
sns.violinplot(x='gender', y='depression_score', hue = 'gender', data=filtro2, palette = 'Set2', legend = False)
```

##### A partir del análisis se observa que más del 60% de las personas en cada género presentan niveles de depresión severa o moderada. Los porcentajes son muy similares entre los distintos grupos (oscilan entre ~61% y ~66%).

#### ESTILO DE VIDA: Nivel de estrés, horas de sueño y días de actividad física.

##### Los rangos que a continuación voy a establecer fueron definidos en base a la distribución de los datos (mediana y percentiles), y alineados con criterios clínicos y recomendaciones generales sobre salud mental, sueño y actividad física.

```python
# Rangos: Niveles de estrés

rangos_stress = [0, 3, 6, 10]  
categorias_s = ["Bajo", "Medio", "Alto"]

df['stress_range'] = pd.cut(df['stress_level'], bins=rangos_stress, labels=categorias_s, include_lowest=True)
```

```python
# Rangos: Hs de sueño

rangos_sleep = [0, 4, 6, 10]  
categorias_sl = ["Malo", "Regular", "Bueno"]

df['sleep_range'] = pd.cut(df['sleep_hours'], bins=rangos_sleep, labels=categorias_sl, include_lowest=True)
```

```python
# Rangos: Días de actividad física

rangos_ac = [0, 1, 3, 7]  
categorias_ac = ["Nula", "Buena", "Excelente"]

df['activity_range'] = pd.cut(df['physical_activity_days'], bins=rangos_ac, labels=categorias_ac, include_lowest=True)
```

#### _Relación ANSIEDAD y ESTILO DE VIDA_

```python
## Gráficos

fig, axes = plt.subplots(1, 3, figsize=(10, 5), sharey=True)

sns.countplot(data=df, x='activity_range', hue='anxiety_level', palette='Set2', ax=axes[0])
axes[0].set_title('Niveles de ansiedad - act física')
axes[0].set_xlabel('activity_range')
axes[0].set_ylabel('anxiety_level')
axes[0].tick_params(axis='x', rotation=45)
axes[0].legend(title='Tipo de ansiedad')


sns.countplot(data=df, x='sleep_range', hue='anxiety_level', palette='Set2',legend = False, ax=axes[1])
axes[1].set_title('Niveles de ansiedad - hs de sueño')
axes[1].set_xlabel('anxiety_level')
axes[1].tick_params(axis='x', rotation=45)

sns.countplot(data=df, x='stress_range', hue='anxiety_level', palette='Set2', legend = False, ax=axes[2])
axes[2].set_title('Niveles de ansiedad - estrés')
axes[2].set_xlabel('anxiety_level')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

#### _Relación DEPRESIÓN y ESTILO DE VIDA_

#### 10. CONCLUSIÓN 

##### El _punto de partida_ de este análisis se enfocó en los altos niveles de ansiedad y depresión observados en la muestra, usando como referencia los test: "GAD-7" para la ansiedad y "PHQ-9" para la depresión. 

##### De la muestra analizada, un 32,06% presenta síntomas compatibles con una _ansiedad_ severa y un 23,27% con una ansiedad moderada, por lo tanto, podemos decir que un **55,33% padece los síntomas de este trastorno.**

##### En cuanto a la _depresión_ los datos fueron aún más llamativos ya que un 35,58% padecen síntomas compatibles con una depresión severa, un 15,81% con una depresión moderada y un 16,61% con una de depresión severamente moderada, concluyendo que del total de la muestra un **68% padece de estos síntomas de manera intensa.**

##### Con el despliegue realizado puedo concluir que la **edad** aparece como la variable **más influyente** en los niveles de ansiedad y depresión.
